```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on= 'Country Code', how ='left')
```

```python
final_df.head(5)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val= final_df.Country.value_counts().values
```

```python
## Pie Chart - Top3 ciuntries that uses Zomato 
plt.pie(country_val[:3], labels = country_names[:3],autopct = '%1.2f%%')
```

Observation:
Zomato maximum records or transactions are from India . After India, there is USA and then UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

Observations:
1. When Rating is between 4.5 to 4.9  ----> Excellent 
2. When Rating is between 4.0 to 4.4  ----> Very good
3. When Rating is between 3.5 to 3.9  ----> Good
4. When Rating is between 2.5 to 3.4  ----> Average
5. When Rating is between 1.8 to 2.4  ----> Poor
6. When Rating is 0.0                 ----> Not rated  

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12, 6)
sns.barplot(x = 'Aggregate rating', y = 'Rating count', data = ratings)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating count', hue = 'Rating color', data = ratings, palette = ['pink','red','orange','yellow','green','green'])
```

Observation:
1. Not rated count is very high
2. maximum number of rating are between 2.5 to 3.4

```python
## count plot 
sns.countplot(x = 'Rating color', data =ratings,palette = ['pink','red','orange','yellow','green','green'])
```

```python
## Number of countries that has given 0 rating
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

Observation: Maximum no of 0 rating are from India
